```python
import pandas as pd
```

```python
sad_df = pd.read_csv('./results/슬픔_topic_labeled.csv')
happy_df = pd.read_csv('./results/기쁨_topic_labeld.csv')
angry_df = pd.read_csv('./results/분노_topic_labeled.csv')
insecure_df = pd.read_csv('./results/불안_topic_labeld.csv')
```

```python
def topic_cnt(feeling, df):
    print(feeling)
    topic1 = df[df['topic'] == 'Topic #1']
    print(f'topic1_cnt: {topic1.shape[0]}개')
    topic2 = df[df['topic'] == 'Topic #2']
    print(f'topic2_cnt: {topic2.shape[0]}개')
    topic3 = df[df['topic'] == 'Topic #3']
    print(f'topic2_cnt: {topic3.shape[0]}개\n')
```

```python
topic_cnt('슬픔', sad_df)
topic_cnt('기쁨', happy_df)
topic_cnt('분노', angry_df)
topic_cnt('불안', insecure_df)
```

```python
topic1 = sad_df[sad_df['topic'] == 'Topic #1']
topic1.sort_values('probability', ascending=False)
```

```python
def top_5_by_topic(df):
    return df.sort_values(by='probability', ascending=False).head(5)
```

```python
sad_top_5 = sad_df.groupby(['topic', 'book_type'], group_keys=False).apply(top_5_by_topic)
happy_top_5 = happy_df.groupby(['topic', 'book_type'], group_keys=False).apply(top_5_by_topic)
insecure_top_5 = insecure_df.groupby(['topic', 'book_type'], group_keys=False).apply(top_5_by_topic)
angry_top_5 = angry_df.groupby(['topic', 'book_type'], group_keys=False).apply(top_5_by_topic)
```

```python
merged_results = pd.concat([sad_top_5, happy_top_5, insecure_top_5, angry_top_5], ignore_index=True)
merged_results
```

```python
merged_results.to_csv('감정별_토픽별_작품형식별_도서리스트.csv', index=False)
```
